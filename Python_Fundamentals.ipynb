Q5. Sum of First N Natural Numbers
· Takes a number N from the user.

· Calculates and prints the sum of first N natural numbers using a loop.

```python
N = int(input("Enter a number: "))

if N < 0:
    print("Please enter a positive number.")
else:
    total_sum = 0
    for i in range(1, N + 1):
        total_sum += i #total_sum = i + 1 gives N+1 as the result
    print("The sum of the first", N, "natural numbers is:", total_sum)
```
